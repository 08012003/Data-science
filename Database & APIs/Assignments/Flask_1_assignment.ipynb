## Q1. What is Flask Framework? What are the advantages of Flask Framework?

Flask is a lightweight and flexible web framework for Python. It's classified as a "microframework" because it provides only the essential tools and libraries needed for building web applications, leaving developers free to choose other components as needed. This minimalist approach gives developers a high degree of control and customization.

### Advantages of Flask Framework

*   **Lightweight & Flexible:** Easy to learn and use, especially for smaller projects and prototypes.
*   **Easy to Learn:** Simple API and clear documentation.
*   **Extensible:** Functionality can be added via extensions.
*   **Control & Customization:** Fine-grained control over application architecture.
*   **Built-in Server & Debugger:** Simplifies development and testing.
*   **Jinja2 Templating:** Enables dynamic HTML page creation.

## Q3. What is App routing in Flask? Why do we use app routes?

In Flask, app routing refers to the process of mapping URLs (Uniform Resource Locators) to specific functions within your Python code. It determines which function should be executed when a user visits a particular URL in their web browser.

The `@app.route()` decorator is used to define these routes. It associates a URL pattern with a function. When Flask receives a request for a URL that matches a defined route, it executes the corresponding function and returns the result to the client (usually a web browser).

### Why do we use App Routes?

App routing is essential for creating dynamic web applications with multiple pages or functionalities. Here's why we use it:

*   **Organization and Structure:** Routing helps organize your application's code by separating different functionalities into distinct functions, each responsible for handling a specific URL. This makes your code more modular, maintainable, and easier to understand.
*   **User Experience:** Routing allows users to navigate to different parts of your application by using different URLs. This provides a clear and intuitive user experience.
*   **URL Management:** Routing gives you control over the structure and appearance of your application's URLs. You can create user-friendly and SEO-friendly URLs.

## Q4. Create a “/welcome” route to display the welcome message “Welcome to ABC Corporation” and a “/”
#### ABC Corporation

*   **Company Name:** ABC Corporation
*   **Location:** India
*   **Contact Detail:** 999-999-9999

```python
from flask import Flask

app = Flask(__name__)

@app.route("/welcome")
def welcome():
    return "Welcome to ABC Corporation"

@app.route("/")
def index():
    return """
    Company Name: ABC Corporation<br>
    Location: India<br>
    Contact Detail: 999-999-9999
    """

if __name__ == "__main__":
    app.run(debug=True)
```

## Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.

```python
"""In Flask, the url_for() function is used for URL building. It generates the URL for a specific function,
which is typically used to create links to specific routes in a Flask web application."""

from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def home():
    return 'Welcome to the Home Page!'

@app.route('/profile')
def profile():
    return 'This is your Profile!'

@app.route('/redirect-to-profile')
def redirect_to_profile():
    profile_url = url_for('profile')  # Generate the URL for the profile route
    return f'Go to the profile page at: <a href="{profile_url}">Profile</a>'

if __name__ == '__main__':
    app.run(debug=True)
```
