```python
import csv
```

```python
data = [["name" , "email_id" , "phone_number"],
        ["sudh" ,"sudh@gmail.com" , 92342424],
        ["krish" , "krish@gmail.com" , 9324242]
]
```

```python
with open("data.csv" , "w" ) as f :
    writer = csv.writer(f)

    for i in data:
        writer.writerow(i)

```

```python
with open("data.csv" , 'r') as f :
    read_data = csv.reader(f)

    for i in read_data:
        print(i)
```
